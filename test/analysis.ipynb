```python
import json
import matplotlib.pyplot as plt
with open('results_analysis.json') as f:
    results = json.load(f)
# Define metrics and datasets
metrics = ['PSNR', 'SSIM', 'L1_Loss']
# datasets = ['Set5', 'Set14'] #, 'Urban100','BSD100'
datasets = ['BSD100']
# Create subplots - one row for each dataset, one column for each metric
fig, axs = plt.subplots(len(datasets), len(metrics), figsize=(15, 5 * len(datasets)))

for dataset_idx, dataset in enumerate(datasets):
    for metric_idx, metric in enumerate(metrics):
        if dataset_idx == 50:
            break
        ax = axs[dataset_idx][metric_idx] if len(datasets) > 1 else axs[metric_idx]

        for model in results:
            x = list(results[model][dataset].keys())  # Image IDs
            y = [results[model][dataset][img_id][metric] for img_id in x]  # Metric values
            ax.plot(x, y, marker='o', label=model, markersize=3)  # Plotting line for each model

        ax.set_title(f'{metric} - {dataset}')
        ax.set_xlabel('Image ID')
        ax.set_ylabel(metric)
        ax.legend()

plt.tight_layout()
plt.show()
```

```python

import matplotlib.pyplot as plt
from PIL import Image

# Results data with PSNR and SSIM values for different models

# Image paths for low-resolution and high-resolution images
# Assuming you have a dictionary for paths; if not, you can directly load the images if you have them
image_paths = {
    'Orginal Low Resolution': 'images/img_002_SRF_4_LR.png',
    
    'srgan_model': 'images/srgan_enhanced.jpg',
    'srflow_model': 'images/srflow_enhanced.jpg',
    'rcan_model': 'images/rcan_enhanced.jpg',
    'hat_model': 'images/hat_enhanced (2).jpg',
    'Original High Resolution': 'images/img_002_SRF_4_HR.png',
}

# Function to load image and get PSNR/SSIM from the results
def load_image_and_metrics(model_name, dataset, image_id):
    image_path = image_paths[model_name]
    image = Image.open(image_path)
    metrics = results[model_name][dataset][image_id]
    psnr_ssim = f"{metrics['PSNR']:.2f}/{metrics['SSIM']:.4f}"
    return image, psnr_ssim

# Assuming you have a dataset name and image ID to look up
dataset = 'Set5'
image_id = '1'

# image_id = '1'
# Create subplots
num_models = len(image_paths)
fig, axes = plt.subplots(1, num_models, figsize=(20, 4))

# Load images and display them with metrics
for i, (model_name, path) in enumerate(image_paths.items()):
    if model_name == 'Orginal Low Resolution' or model_name == 'Original High Resolution':
        image = Image.open(path)
        axes[i].imshow(image)
        axes[i].set_title(model_name)
        axes[i].axis('off')
    else:
        image, psnr_ssim = load_image_and_metrics(model_name, dataset, image_id)
        
        axes[i].imshow(image)
        axes[i].set_title(f"{model_name}\nPSNR/SSIM: {psnr_ssim}")
        axes[i].axis('off')  # Hide axes

plt.tight_layout()
plt.show()
```
